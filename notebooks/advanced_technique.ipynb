```python
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Stroke_Advanced") \
    .master("local[*]") \
    .getOrCreate()
```

```python
train = spark.read.parquet("../data/train_gold.parquet")
test = spark.read.parquet("../data/test_gold.parquet")

with open("../data/feature_metadata.json", "r", encoding="utf-8") as f:
    feature_meta = json.load(f)

idx_to_label = {}
for feat in feature_meta["features"]:
    if feat["type"] == "onehot":
        idx_to_label[feat["index"]] = f"{feat['column']}_{feat['category']}"
    else:
        idx_to_label[feat["index"]] = feat["column"]

print(f"Train: {train.count()}, Test: {test.count()}, Features: {len(idx_to_label)}")
```
